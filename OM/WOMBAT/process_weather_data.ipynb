```python
import pandas as pd  
import numpy as np
from datetime import datetime, timedelta
```

```python
data = pd.read_pickle("cell_profile_with_typhoon.pkl")
data 
```

```python

highest_cap_cell = data.loc[data['wind_offshore_fixed_cap'].idxmax()]
print(highest_cap_cell)

wind_speeds = highest_cap_cell['wind_speed_typhoon']  

dates = []
start_year = 2002
end_year = 2021
for year in range(start_year, end_year + 1):
    start_date = datetime(year, 1, 1)
    for hour in range(8760):
        current_date = start_date + timedelta(hours=hour)
        if current_date.year == year: 
            dates.append(current_date.strftime("%m/%d/%y %H:%M"))

all_wind_speeds = np.tile(wind_speeds, end_year - start_year + 1)

result_df = pd.DataFrame({
    'datetime': dates,
    'windspeed': all_wind_speeds
})
result_df.head()
```

```python
result_df.to_csv("library/weather/weather.csv", index=False)
```
